# 🔍 RAG 검색 디버깅 노트북

벡터 검색 + Cohere Reranking 파이프라인 테스트

**테스트 항목**:
- Dense 벡터 검색 (Gemini embedding)
- Sparse 벡터 검색 (BM25 via FastEmbed)
- **하이브리드 검색 (Dense 0.7 + Sparse 0.3)**
- Cohere 리랭킹
- 콘텐츠 타입 (text/table/list)
- 테이블 요약 검색

**실행 순서**: 모든 셀을 위에서 아래로 순차적으로 실행

> ⚠️ 하이브리드 검색을 위해 컬렉션 재생성 및 문서 리인덱싱 필요:
> 1. 컬렉션 재생성 (sparse vector config 포함)
> 2. `curl -X POST http://localhost:8000/documents/reindex/{document_id}`

````python
# 환경 설정
import os
import sys
from pathlib import Path

# .env 로드 (반드시 먼저 실행)
from dotenv import load_dotenv
load_dotenv()

# API 키 확인
print("환경변수 상태")
print("=" * 50)
keys = ["GEMINI_API_KEY", "COHERE_API_KEY", "QDRANT_HOST", "QDRANT_PORT"]
all_ok = True
for key in keys:
    val = os.getenv(key)
    status = "✓" if val else "✗ MISSING"
    if not val:
        all_ok = False
    display = val[:20] + "..." if val and len(val) > 20 else (val or "")
    print(f"  {key}: {status} ({display})")

if not all_ok:
    print("\n⚠️ 누락된 환경변수가 있습니다. .env 파일을 확인하세요.")
````

````python
# document.py 함수 임포트
from routers.document import (
    get_reranker,
    get_sparse_embeddings,
    QDRANT_HOST,
    QDRANT_PORT,
    COLLECTION_NAME,
    DENSE_WEIGHT,
    SPARSE_WEIGHT,
)

print(f"Qdrant: {QDRANT_HOST}:{QDRANT_PORT}")
print(f"컬렉션: {COLLECTION_NAME}")
print(f"하이브리드 가중치: Dense={DENSE_WEIGHT}, Sparse={SPARSE_WEIGHT}")
````

````python
# Qdrant 클라이언트 초기화
from qdrant_client import QdrantClient

qdrant = QdrantClient(host=QDRANT_HOST, port=QDRANT_PORT)

# 컬렉션 상태 확인
collections = qdrant.get_collections().collections
print("📦 Qdrant 컬렉션 목록")
print("=" * 50)
for c in collections:
    info = qdrant.get_collection(c.name)
    print(f"  {c.name}: {info.points_count} points")
````

````python
# 검색 쿼리 설정
SEARCH_QUERY = "LAM(Large Active Model)이란 무엇인가요?"  # 🎯 테스트할 쿼리
TOP_K_VECTOR = 30  # 1차 벡터 검색 개수
TOP_K_RERANK = 5   # 2차 리랭킹 후 최종 개수

print(f"🔍 검색 쿼리: '{SEARCH_QUERY}'")
print(f"   1차 벡터: {TOP_K_VECTOR}개 → 2차 리랭크: {TOP_K_RERANK}개")
````

````python
# Step 1: 쿼리 임베딩 생성
from google import genai

# API 키 검증 (dotenv 로드 확인)
api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise ValueError(
        "GEMINI_API_KEY가 설정되지 않았습니다.\n"
        "1. .env 파일에 GEMINI_API_KEY가 있는지 확인\n"
        "2. 위의 '환경 설정' 셀을 먼저 실행했는지 확인"
    )

gemini_client = genai.Client(api_key=api_key)
query_embedding = gemini_client.models.embed_content(
    model="text-embedding-004",
    contents=SEARCH_QUERY,
).embeddings[0].values

print(f"✅ 쿼리 임베딩 생성 완료")
print(f"   Dimension: {len(query_embedding)}")
print(f"   Sample: {query_embedding[:5]}")
````

````python
# Step 1.5: Sparse 쿼리 임베딩 생성 (BM25)
sparse_result = get_sparse_embeddings([SEARCH_QUERY])
query_sparse = sparse_result[0]  # (indices, values)

print(f"✅ Sparse 쿼리 임베딩 생성 완료")
print(f"   Non-zero dimensions: {len(query_sparse[0])}")
print(f"   Top 10 terms (by weight):")

# 가중치 높은 순으로 정렬하여 표시
sorted_terms = sorted(zip(query_sparse[0], query_sparse[1]), key=lambda x: -x[1])[:10]
for idx, (term_idx, weight) in enumerate(sorted_terms):
    print(f"     {idx+1}. term_idx={term_idx}, weight={weight:.4f}")
````

````python
# Step 2: 벡터 검색
vector_results = qdrant.query_points(
    collection_name=COLLECTION_NAME,
    query=query_embedding,
    using="dense",
    limit=TOP_K_VECTOR,
    with_payload=True,
)

print(f"📊 1차 벡터 검색 결과: {len(vector_results.points)}개\n")
print(f"{'Rank':<5} {'Score':<8} {'Type':<6} {'Heading':<45}")
print("=" * 70)

for i, hit in enumerate(vector_results.points):
    heading = hit.payload.get('heading', 'N/A')[:43]
    content_type = hit.payload.get('content_type', 'text')[:5]
    print(f"{i+1:<5} {hit.score:<8.4f} {content_type:<6} {heading}")
````

````python
# Step 2.5: Sparse 벡터 검색
from qdrant_client.models import SparseVector

sparse_search_result = qdrant.query_points(
    collection_name=COLLECTION_NAME,
    query=SparseVector(indices=query_sparse[0], values=query_sparse[1]),
    using="sparse",
    limit=TOP_K_VECTOR,
    with_payload=True,
)
sparse_results = sparse_search_result.points

print(f"📊 Sparse 벡터 검색 결과: {len(sparse_results)}개\n")
print(f"{'Rank':<5} {'Score':<8} {'Type':<6} {'Heading':<45}")
print("=" * 70)

for i, hit in enumerate(sparse_results):
    heading = hit.payload.get('heading', 'N/A')[:43]
    content_type = hit.payload.get('content_type', 'text')[:5]
    print(f"{i+1:<5} {hit.score:<8.4f} {content_type:<6} {heading}")
````

````python
# Step 2.7: 하이브리드 스코어 계산 및 비교
print(f"📊 하이브리드 스코어 계산 (Dense={DENSE_WEIGHT}, Sparse={SPARSE_WEIGHT})\n")

# Dense 결과를 딕셔너리로 변환
combined = {}
for r in vector_results.points:
    combined[r.id] = {
        "dense_score": r.score,
        "sparse_score": 0.0,
        "payload": r.payload
    }

# Sparse 결과 병합
for r in sparse_results:
    if r.id in combined:
        combined[r.id]["sparse_score"] = r.score
    else:
        combined[r.id] = {
            "dense_score": 0.0,
            "sparse_score": r.score,
            "payload": r.payload
        }

# 하이브리드 스코어 계산
for data in combined.values():
    data["hybrid_score"] = DENSE_WEIGHT * data["dense_score"] + SPARSE_WEIGHT * data["sparse_score"]

# 정렬
sorted_results = sorted(combined.items(), key=lambda x: x[1]["hybrid_score"], reverse=True)

# 비교 테이블 출력
print(f"{'Rank':<5} {'Hybrid':<9} {'Dense':<9} {'Sparse':<9} {'Type':<6} {'Heading':<35}")
print("=" * 85)

for i, (point_id, data) in enumerate(sorted_results[:15]):
    heading = data["payload"].get('heading', 'N/A')[:33] if data["payload"] else "N/A"
    content_type = data["payload"].get('content_type', 'text')[:5] if data["payload"] else "text"
    print(f"{i+1:<5} {data['hybrid_score']:<9.4f} {data['dense_score']:<9.4f} {data['sparse_score']:<9.4f} {content_type:<6} {heading}")
````

````python
# Step 2.8: Dense-only vs Hybrid 순위 변화 분석
print("📈 Dense-only vs Hybrid 순위 변화 분석\n")

# Dense-only 순위 (원래 벡터 검색 결과)
dense_ranks = {hit.id: i+1 for i, hit in enumerate(vector_results.points)}

# Hybrid 순위
hybrid_ranks = {point_id: i+1 for i, (point_id, _) in enumerate(sorted_results)}

# 순위 변화가 큰 항목들 찾기
rank_changes = []
for point_id, data in combined.items():
    dense_rank = dense_ranks.get(point_id, 999)
    hybrid_rank = hybrid_ranks.get(point_id, 999)
    change = dense_rank - hybrid_rank  # 양수면 순위 상승
    
    if dense_rank <= 30 or hybrid_rank <= 30:  # 상위 30위 내 항목만
        rank_changes.append({
            "point_id": point_id,
            "heading": data["payload"].get("heading", "N/A")[:40] if data["payload"] else "N/A",
            "dense_rank": dense_rank,
            "hybrid_rank": hybrid_rank,
            "change": change,
            "dense_score": data["dense_score"],
            "sparse_score": data["sparse_score"],
        })

# 순위 상승/하락 분류
print("🔼 Hybrid에서 순위 상승한 문서 (Sparse 검색의 기여)")
print("-" * 80)
rank_ups = sorted([r for r in rank_changes if r["change"] > 0], key=lambda x: -x["change"])[:10]
for r in rank_ups:
    print(f"  Dense #{r['dense_rank']:2d} → Hybrid #{r['hybrid_rank']:2d} (+{r['change']:2d}) | sparse={r['sparse_score']:.4f} | {r['heading']}")

print("\n🔽 Hybrid에서 순위 하락한 문서 (Dense 점수 희석)")
print("-" * 80)
rank_downs = sorted([r for r in rank_changes if r["change"] < 0], key=lambda x: x["change"])[:10]
for r in rank_downs:
    print(f"  Dense #{r['dense_rank']:2d} → Hybrid #{r['hybrid_rank']:2d} ({r['change']:2d}) | sparse={r['sparse_score']:.4f} | {r['heading']}")
````

````python
# Step 3: Cohere Reranking (하이브리드 결과 기반)
from langchain_core.documents import Document

reranker = get_reranker(top_n=TOP_K_RERANK)

# 하이브리드 상위 20개를 리랭킹 대상으로
docs_for_rerank = [
    Document(
        page_content=data["payload"].get("content", "") if data["payload"] else "",
        metadata={
            "chunk_id": data["payload"].get("chunk_id") if data["payload"] else None,
            "heading": data["payload"].get("heading") if data["payload"] else None,
            "hierarchy_path": data["payload"].get("hierarchy_path") if data["payload"] else None,
            "dense_score": data["dense_score"],
            "sparse_score": data["sparse_score"],
            "hybrid_score": data["hybrid_score"],
            # 콘텐츠 타입 필드
            "content_type": data["payload"].get("content_type", "text") if data["payload"] else "text",
            "table_summary": data["payload"].get("table_summary") if data["payload"] else None,
            "has_context": data["payload"].get("has_context", False) if data["payload"] else False,
            "table_context": data["payload"].get("table_context") if data["payload"] else None,
        }
    )
    for point_id, data in sorted_results[:20]
]

reranked_docs = reranker.compress_documents(docs_for_rerank, SEARCH_QUERY)

print(f"📊 Cohere 리랭킹 결과 (하이브리드 기반): {len(reranked_docs)}개\n")
print(f"{'Rank':<5} {'Rerank':<8} {'Hybrid':<8} {'Dense':<8} {'Sparse':<8} {'Type':<6} {'Heading':<30}")
print("=" * 90)

for i, doc in enumerate(reranked_docs):
    rerank_score = doc.metadata.get('relevance_score', 0)
    hybrid_score = doc.metadata.get('hybrid_score', 0)
    dense_score = doc.metadata.get('dense_score', 0)
    sparse_score = doc.metadata.get('sparse_score', 0)
    content_type = doc.metadata.get('content_type', 'text')[:5]
    heading = doc.metadata.get('heading', 'N/A')[:28]
    print(f"{i+1:<5} {rerank_score:<8.4f} {hybrid_score:<8.4f} {dense_score:<8.4f} {sparse_score:<8.4f} {content_type:<6} {heading}")
````

````python
# Hybrid vs Rerank 순위 변화 분석
print("📈 Hybrid → Rerank 순위 변화 분석\n")

# Hybrid 순위 매핑 (chunk_id 기준)
hybrid_chunk_ranks = {}
for i, (point_id, data) in enumerate(sorted_results):
    chunk_id = data["payload"].get("chunk_id") if data["payload"] else None
    if chunk_id:
        hybrid_chunk_ranks[chunk_id] = i + 1

print(f"{'Final':<6} {'Hybrid':<7} {'Change':<8} {'Heading':<50}")
print("=" * 75)

for i, doc in enumerate(reranked_docs):
    chunk_id = doc.metadata.get('chunk_id')
    hybrid_rank = hybrid_chunk_ranks.get(chunk_id, '?')
    change = hybrid_rank - (i + 1) if isinstance(hybrid_rank, int) else 0
    change_str = f"+{change}" if change > 0 else str(change)
    heading = doc.metadata.get('heading', 'N/A')[:48]
    
    print(f"{i+1:<6} {hybrid_rank:<7} {change_str:<8} {heading}")
````

````python
# 최종 결과 상세 보기 (하이브리드 + 리랭킹)
if reranked_docs:
    for idx, doc in enumerate(reranked_docs):
        print(f"📄 검색 결과 #{idx + 1} 상세")
        print("=" * 60)
        print(f"Heading: {doc.metadata.get('heading')}")
        print(f"Path: {doc.metadata.get('hierarchy_path')}")
        
        # 스코어 비교
        print(f"\n--- 스코어 비교 ---")
        print(f"Rerank Score:  {doc.metadata.get('relevance_score', 0):.4f}")
        print(f"Hybrid Score:  {doc.metadata.get('hybrid_score', 0):.4f}")
        print(f"  ├─ Dense:    {doc.metadata.get('dense_score', 0):.4f} × {DENSE_WEIGHT}")
        print(f"  └─ Sparse:   {doc.metadata.get('sparse_score', 0):.4f} × {SPARSE_WEIGHT}")
        
        # 콘텐츠 타입 정보
        print(f"\n--- 콘텐츠 타입 정보 ---")
        print(f"Content Type: {doc.metadata.get('content_type', 'text')}")
        print(f"Has Context: {doc.metadata.get('has_context', False)}")
        
        # 테이블인 경우 추가 정보
        if doc.metadata.get('content_type') == 'table':
            table_context = doc.metadata.get('table_context')
            table_summary = doc.metadata.get('table_summary')
            
            if table_context:
                print(f"\n--- 테이블 컨텍스트 (앞 설명문) ---")
                print(table_context)
            
            if table_summary:
                print(f"\n--- 테이블 요약 (LLM 생성) ---")
                print(table_summary)
        
        print(f"\n--- Content ---\n")
        print(doc.page_content)
        print("\n" + "─" * 60 + "\n")
else:
    print("검색 결과가 없습니다.")
````

````python
# 콘텐츠 타입별 통계
from collections import Counter

print("📊 검색 결과 콘텐츠 타입 분포\n")

# 벡터 검색 결과 통계
vector_types = Counter(hit.payload.get('content_type', 'text') for hit in vector_results.points)
print("1차 벡터 검색 (Top 20):")
for ctype, count in vector_types.most_common():
    print(f"  {ctype}: {count}개")

# 리랭킹 결과 통계
rerank_types = Counter(doc.metadata.get('content_type', 'text') for doc in reranked_docs)
print(f"\n2차 리랭킹 (Top {TOP_K_RERANK}):")
for ctype, count in rerank_types.most_common():
    print(f"  {ctype}: {count}개")

# 테이블 요약 유무 확인
tables_with_summary = sum(1 for doc in reranked_docs 
                          if doc.metadata.get('content_type') == 'table' 
                          and doc.metadata.get('table_summary'))
tables_total = sum(1 for doc in reranked_docs if doc.metadata.get('content_type') == 'table')

if tables_total > 0:
    print(f"\n📋 테이블 요약 현황: {tables_with_summary}/{tables_total}개 (리인덱싱 필요시 요약 없음)")
````

# 🔄 Critique & Filter Output 테스트

LangGraph Agent를 통한 전체 파이프라인 테스트:
- **Critique**: 응답 품질 평가 및 재생성 사이클
- **Filter Output**: `user_facing=True` 메시지만 필터링

> ⚠️ **전제조건**: LangGraph 서버 실행 필요
> ```bash
> cd rag-agent && langgraph dev
> ```

````python
# LangGraph SDK 연결
from langgraph_sdk import get_client

# LangGraph 서버 연결 (기본 포트: 2024)
LANGGRAPH_URL = "http://localhost:2024"
ASSISTANT_ID = "chatbot"  # langgraph.json의 graph 이름

client = get_client(url=LANGGRAPH_URL)

# 연결 테스트: assistants 목록 조회
try:
    assistants = await client.assistants.search()
    print(f"✅ LangGraph 서버 연결 성공: {LANGGRAPH_URL}")
    print(f"   사용 가능한 Assistants:")
    for a in assistants:
        print(f"   - {a['assistant_id']} (graph: {a.get('graph_id', 'N/A')})")
except Exception as e:
    print(f"❌ LangGraph 서버 연결 실패: {e}")
    print(f"   → 'cd rag-agent && langgraph dev' 실행 필요")
````

````python
# Agent 호출 테스트 (Critique 포함 전체 파이프라인)
print(f"🔍 테스트 쿼리: '{SEARCH_QUERY}'")
print("=" * 60)

# 새 스레드 생성 및 agent 호출
thread = await client.threads.create()
print(f"📝 Thread 생성: {thread['thread_id']}")

# Agent 실행 (검색 → 응답 생성 → Critique → Filter Output)
print("\n⏳ Agent 실행 중 (검색 + Critique 사이클)...")
response = await client.runs.wait(
    thread_id=thread["thread_id"],
    assistant_id=ASSISTANT_ID,
    input={"messages": [{"role": "user", "content": SEARCH_QUERY}]},
)

print(f"✅ Agent 실행 완료")
print(f"   Run ID: {response.get('run_id', 'N/A')}")
print(f"   Status: {response.get('status', 'N/A')}")
````

````python
# Critique 결과 확인
state = await client.threads.get_state(thread["thread_id"])
values = state["values"]

print("📊 Critique 분석 결과")
print("=" * 60)

# Critique 관련 상태 출력
critique_count = values.get("critique_count", 0)
critique_feedback = values.get("critique_feedback", "없음")

print(f"🔄 Critique 횟수: {critique_count}")
print(f"   (최대 3회까지 재생성 시도)")

if critique_feedback and critique_feedback != "없음":
    print(f"\n📝 마지막 Critique 피드백:")
    print(f"   {critique_feedback}")
else:
    print(f"\n✅ 첫 응답이 기준을 통과했거나 피드백 없음")

# 추가 상태 정보 (있을 경우)
if "critique_verdict" in values:
    print(f"\n🏷️  Critique Verdict: {values['critique_verdict']}")
if "critique_score" in values:
    print(f"📈 Critique Score: {values['critique_score']}")
````

````python
# Filter Output 결과 확인 (user_facing 메시지만)
messages = values.get("messages", [])

print("📤 Filter Output 분석")
print("=" * 60)

# 전체 메시지 vs user_facing 메시지
total_messages = len(messages)
user_facing_messages = [
    m for m in messages
    if m.get("additional_kwargs", {}).get("user_facing") is True
]

print(f"📊 메시지 통계:")
print(f"   전체 메시지: {total_messages}개")
print(f"   user_facing: {len(user_facing_messages)}개")
print(f"   필터링된 내부 메시지: {total_messages - len(user_facing_messages)}개")

# 최종 사용자 응답 출력
if user_facing_messages:
    final_response = user_facing_messages[-1]
    print(f"\n🎯 최종 사용자 응답 (user_facing=True):")
    print("-" * 60)
    content = final_response.get("content", "")
    # 긴 응답은 앞부분만 표시
    if len(content) > 1000:
        print(content[:1000])
        print(f"\n... (총 {len(content)}자, 이하 생략)")
    else:
        print(content)
else:
    print("\n⚠️ user_facing 메시지가 없습니다.")
    print("   filter_output 노드가 정상 동작하는지 확인 필요")
````

````python
# 전체 메시지 히스토리 상세 (디버깅용)
print("📜 전체 메시지 히스토리 (내부 처리 포함)")
print("=" * 60)

for idx, msg in enumerate(messages):
    role = msg.get("type", msg.get("role", "unknown"))
    user_facing = msg.get("additional_kwargs", {}).get("user_facing", False)
    content = msg.get("content", "")
    
    # 역할별 이모지
    emoji = {
        "human": "👤",
        "ai": "🤖", 
        "assistant": "🤖",
        "tool": "🔧",
        "system": "⚙️"
    }.get(role, "❓")
    
    # user_facing 표시
    facing_badge = "✅ user_facing" if user_facing else "🔒 internal"
    
    # 콘텐츠 미리보기 (100자)
    preview = content[:100].replace('\n', ' ') if content else "(empty)"
    if len(content) > 100:
        preview += "..."
    
    print(f"\n{idx+1}. {emoji} [{role}] {facing_badge}")
    print(f"   {preview}")
````
