# RAGAS 평가 노트북

RAG 파이프라인의 품질을 정량적으로 평가하는 노트북

**평가 메트릭**:
- **Faithfulness**: 응답이 검색된 컨텍스트에 근거하는지 (환각 탐지)
- **Answer Relevancy**: 응답이 질문에 적절한지
- **Context Precision**: 검색된 컨텍스트 중 관련 있는 비율
- **Context Recall**: 정답 생성에 필요한 컨텍스트가 충분히 검색되었는지

**Judge LLM**: `gpt-4o-mini`

**실행 전제조건**:
1. `uv sync` 로 ragas 설치
2. Qdrant 서버 실행 중 (localhost:6333)
3. `langgraph dev` 로 LangGraph 서버 실행 (localhost:2024)
4. `.env`에 `OPENAI_API_KEY` 설정

## 2-1. 환경 설정

```python
import os
import sys
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()
sys.path.insert(0, str(Path.cwd() / "src"))

# RAGAS imports (v0.4+)
from ragas import evaluate, EvaluationDataset, SingleTurnSample
from ragas.llms import llm_factory
from ragas.embeddings import OpenAIEmbeddings as RagasOpenAIEmbeddings
from ragas.metrics.collections import (
    Faithfulness,
    AnswerRelevancy,
    ContextPrecision,
    ContextRecall,
)
from ragas.testset import TestsetGenerator
from openai import AsyncOpenAI
from langgraph_sdk import get_client

# 환경변수 확인
print("환경변수 상태")
print("=" * 50)
for key in ["OPENAI_API_KEY", "QDRANT_HOST", "QDRANT_PORT"]:
    val = os.getenv(key)
    status = "✓" if val else "✗ MISSING"
    display = val[:20] + "..." if val and len(val) > 20 else (val or "")
    print(f"  {key}: {status} ({display})")
```

## 2-2. Judge LLM 설정

```python
openai_client = AsyncOpenAI()
evaluator_llm = llm_factory("gpt-4o-mini", client=openai_client)
evaluator_embeddings = RagasOpenAIEmbeddings(client=openai_client, model="text-embedding-3-small")

print("✅ Judge LLM: gpt-4o-mini")
print("✅ Embeddings: text-embedding-3-small")
```

## 2-3. Qdrant에서 문서 로드 → LangChain Document 변환

```python
from qdrant_client import QdrantClient
from langchain_core.documents import Document
from chatbot.shared.vector_search import get_collection_name, QDRANT_HOST, QDRANT_PORT

COMPANY = "wiseai"
COLLECTION = get_collection_name(COMPANY)

qdrant = QdrantClient(host=QDRANT_HOST, port=QDRANT_PORT)

# Qdrant에서 모든 포인트 가져오기
points = qdrant.scroll(collection_name=COLLECTION, limit=200, with_payload=True)[0]

# LangChain Document로 변환
documents = [
    Document(
        page_content=p.payload.get("content", ""),
        metadata={
            "heading": p.payload.get("heading", ""),
            "hierarchy_path": p.payload.get("hierarchy_path", ""),
            "content_type": p.payload.get("content_type", "text"),
            "original_filename": p.payload.get("original_filename", ""),
        }
    )
    for p in points
    if p.payload.get("content", "").strip()  # 빈 콘텐츠 제외
]

print(f"컬렉션: {COLLECTION}")
print(f"전체 포인트: {len(points)}개")
print(f"로드된 문서 청크: {len(documents)}개 (빈 콘텐츠 제외)")
print(f"\n샘플 문서 (첫 3개):")
for i, doc in enumerate(documents[:3]):
    print(f"  [{i+1}] heading: {doc.metadata['heading'][:50]}")
    print(f"      content: {doc.page_content[:80]}...")
```

## 2-4. 합성 평가 데이터셋 생성 (TestsetGenerator)

RAGAS의 `TestsetGenerator`를 사용하여 Qdrant 문서 기반으로 평가용 질문-답변 쌍을 자동 생성합니다.

- 생성된 데이터셋은 `eval_dataset.json`으로 저장하여 재사용 가능
- 초기 20개 샘플 생성 (비용 ~$0.10)

```python
generator = TestsetGenerator(
    llm=evaluator_llm,
    embedding_model=evaluator_embeddings,
)

testset = generator.generate_with_langchain_docs(
    documents=documents,
    testset_size=20,  # 초기 20개, 이후 확장 가능
)

# 생성된 데이터셋 확인
df_testset = testset.to_pandas()
print(f"생성된 테스트 샘플: {len(df_testset)}개\n")
print(df_testset[["user_input", "reference"]].head(10).to_string(max_colwidth=60))

# JSON으로 저장 (재사용)
df_testset.to_json("eval_dataset.json", orient="records", force_ascii=False, indent=2)
print(f"\n✅ 데이터셋 저장: eval_dataset.json")
```

## 2-5. RAG 파이프라인 실행 및 데이터 수집

LangGraph SDK로 각 테스트 쿼리를 실행하고 상태에서 응답 + 검색 컨텍스트를 추출합니다.

| RAGAS 필드 | LangGraph State 출처 | 설명 |
|-----------|---------------------|------|
| `user_input` | 테스트 쿼리 | 사용자 질문 |
| `response` | `messages[-1].content` (user_facing=True) | 최종 AI 답변 |
| `retrieved_contexts` | `state["values"]["documents"]` | 검색 결과 리스트 |
| `reference` | TestsetGenerator 생성 | Ground truth 답변 |
| `reference_contexts` | TestsetGenerator 생성 | 이상적인 컨텍스트 |

```python
lg_client = get_client(url="http://localhost:2024")

# 연결 테스트
try:
    assistants = await lg_client.assistants.search()
    print(f"✅ LangGraph 서버 연결 성공")
    for a in assistants:
        print(f"   - {a['assistant_id']} (graph: {a.get('graph_id', 'N/A')})")
except Exception as e:
    print(f"❌ LangGraph 서버 연결 실패: {e}")
    print(f"   → 'cd rag-agent && langgraph dev' 실행 필요")
```

```python
async def run_rag(query: str) -> dict:
    """단일 쿼리에 대해 RAG 파이프라인 실행 후 응답 + 컨텍스트 추출."""
    thread = await lg_client.threads.create()
    await lg_client.runs.wait(
        thread_id=thread["thread_id"],
        assistant_id="rag_agent",
        input={"messages": [{"role": "user", "content": query}]},
    )
    state = await lg_client.threads.get_state(thread["thread_id"])
    values = state["values"]

    # 최종 AI 응답 추출 (user_facing=True)
    ai_response = ""
    for msg in reversed(values.get("messages", [])):
        if msg.get("type") == "ai" and msg.get("additional_kwargs", {}).get("user_facing"):
            ai_response = msg.get("content", "")
            break

    return {
        "response": ai_response,
        "retrieved_contexts": values.get("documents", []),
    }


# 전체 데이터셋에 대해 실행
samples = []
total = len(df_testset)

for idx, row in df_testset.iterrows():
    query = row["user_input"]
    print(f"[{idx+1}/{total}] {query[:60]}...", end=" ")

    try:
        result = await run_rag(query)
        samples.append(SingleTurnSample(
            user_input=query,
            response=result["response"],
            retrieved_contexts=result["retrieved_contexts"],
            reference=row.get("reference"),
            reference_contexts=row.get("reference_contexts"),
        ))
        print(f"✅ (응답 {len(result['response'])}자, 컨텍스트 {len(result['retrieved_contexts'])}개)")
    except Exception as e:
        print(f"❌ {e}")

dataset = EvaluationDataset(samples=samples)
print(f"\n수집 완료: {len(samples)}/{total}개 샘플")
```

## 2-6. RAGAS 평가 실행

4가지 핵심 메트릭으로 평가:
- **Faithfulness**: 응답이 컨텍스트에 근거하는가? (0=완전 환각, 1=완벽 근거)
- **Answer Relevancy**: 응답이 질문과 관련있는가? (0=무관, 1=완벽 관련)
- **Context Precision**: 검색 결과 중 관련 문서 비율 (0=전부 무관, 1=전부 관련)
- **Context Recall**: 정답에 필요한 정보가 검색되었는가? (0=미검색, 1=전부 검색)

```python
metrics = [
    Faithfulness(llm=evaluator_llm),
    AnswerRelevancy(llm=evaluator_llm, embeddings=evaluator_embeddings),
    ContextPrecision(llm=evaluator_llm),
    ContextRecall(llm=evaluator_llm),
]

result = evaluate(
    dataset=dataset,
    metrics=metrics,
)

print(result)
```

## 2-7. 결과 분석 및 시각화

```python
df = result.to_pandas()

# 전체 요약
print("=" * 50)
print("RAGAS 평가 결과 요약")
print("=" * 50)

metric_cols = ["faithfulness", "answer_relevancy", "context_precision", "context_recall"]
summary = df[metric_cols].describe()
print(summary)

# 평균 점수 한눈에 보기
print("\n--- 평균 점수 ---")
for m in metric_cols:
    mean = df[m].mean()
    status = "✅" if mean >= 0.7 else "⚠️" if mean >= 0.5 else "❌"
    print(f"  {status} {m}: {mean:.3f}")

# 문제 항목 식별
print("\n--- 환각 의심 항목 (Faithfulness < 0.7) ---")
low = df[df["faithfulness"] < 0.7]
if len(low) == 0:
    print("  없음 ✅")
else:
    for _, row in low.iterrows():
        print(f"  Q: {row['user_input'][:60]}...")
        print(f"     Faithfulness: {row['faithfulness']:.2f}")

print("\n--- 검색 품질 저하 항목 (Context Recall < 0.5) ---")
low_recall = df[df["context_recall"] < 0.5]
if len(low_recall) == 0:
    print("  없음 ✅")
else:
    for _, row in low_recall.iterrows():
        print(f"  Q: {row['user_input'][:60]}...")
        print(f"     Context Recall: {row['context_recall']:.2f}")

# 결과 저장
df.to_json("eval_results.json", orient="records", force_ascii=False, indent=2)
print(f"\n✅ 결과 저장: eval_results.json")
```
